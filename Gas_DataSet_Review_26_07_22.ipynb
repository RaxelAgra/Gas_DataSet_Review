```python
plt.figure(figsize = (4,4))
plt.plot(gas['Year'],gas['Canada'],'or-', label = ['Canada'])
plt.plot(gas['Year'],gas['France'],'og-', label = ['France'])
plt.plot(gas['Year'],gas['Italy'],'om-', label = ['Italy'])
plt.ylabel('Dollar per Gallon')
plt.xlabel('Year')
plt.title('World Gas Prices')
plt.xticks(gas['Year'], rotation = 90)
plt.legend()
plt.savefig('Gasgraph.png')
plt.show()
```
